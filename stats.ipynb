```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/gregoryweber/Documents/cogs-118c-song-familiarity/data/subject1/sub-01_task-songfamiliarity_beh.tsv', sep='\t')
rec = data[data['outcomeMC'] == 1]
notrec = data[data['outcomeMC'] == 0]

plt.figure(figsize=(8, 6))
plt.boxplot([rec['songDur'], notrec['songDur']], labels=['Recognized', 'Not Recognized'])
plt.ylabel('Reaction Time (seconds)')
plt.title('Reaction Times for Recognized vs Not Recognized Songs')
plt.show()
```
